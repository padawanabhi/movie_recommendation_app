```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.impute import SimpleImputer, KNNImputer

from sklearn.decomposition import NMF
import pickle
```

```python
ratings = pd.read_csv('./ratings_matrix_original.csv', index_col=0)
```

```python
constant_imputer = SimpleImputer(strategy='constant', fill_value=0, copy=True)
```

```python
mean_imputer = SimpleImputer(strategy='mean', copy=True)
```

```python
median_imputer = SimpleImputer(strategy='median', copy=True)
```

```python
mode_imputer = SimpleImputer(strategy='most_frequent', copy=True)
```

```python
knn_imputer = KNNImputer(n_neighbors=5, copy=True)
```

```python
ratings.head()
```

```python
constant_imputed_rating = pd.DataFrame(constant_imputer.fit_transform(ratings), 
            index = ratings.index,
            columns = ratings.columns)
```

```python
constant_imputed_rating.head()
```

```python
mean_imputed_rating = pd.DataFrame(mean_imputer.fit_transform(ratings), 
            index = ratings.index,
            columns = ratings.columns)
```

```python
mean_imputed_rating.head()
```

```python
median_imputed_rating = pd.DataFrame(median_imputer.fit_transform(ratings), 
            index = ratings.index,
            columns = ratings.columns)
```

```python
median_imputed_rating.head()
```

```python
mode_imputed_rating = pd.DataFrame(mode_imputer.fit_transform(ratings), 
            index = ratings.index,
            columns = ratings.columns)
```

```python
mode_imputed_rating.head()
```

```python
knn_imputed_rating = pd.DataFrame(knn_imputer.fit_transform(ratings), 
            index = ratings.index,
            columns = ratings.columns)
```

```python
knn_imputed_rating.head()
```

## NMF with constant

```python
n_components = 20
nmf= NMF(n_components=n_components, max_iter=300)
```

```python
nmf.fit(constant_imputed_rating)
```

```python
Q = pd.DataFrame(nmf.components_, columns=list(constant_imputed_rating.columns), 
                 index=[f"cluster_{i+1}" for i in range(n_components)])
```

```python
Q.head(2)
```

```python
P = pd.DataFrame(nmf.transform(constant_imputed_rating),
             index=constant_imputed_rating.index,
            columns = [f"cluster_{i+1}" for i in range(n_components)])
```

```python
P.head(2)
```

```python
R_hat = pd.DataFrame(np.dot(P, Q), 
             index=constant_imputed_rating.index,
            columns=list(constant_imputed_rating.columns))
```

```python
abs(constant_imputed_rating - R_hat)
```

```python
nmf.reconstruction_err_
```
