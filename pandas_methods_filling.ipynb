```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.decomposition import NMF
import pickle
```

```python
ratings = pd.read_csv('./ratings_matrix_original.csv', index_col=0)
```

```python
ratings.head()
```

```python
ratings.T.head()
```

```python
mean_user_ratings = ratings.transpose().fillna(round(ratings.mean(axis=1), 1)).transpose().head()
```

```python
mean_user_ratings.head()
```

```python
median_user_ratings = ratings.transpose().fillna(round(ratings.median(axis=1), 1)).transpose().head()
```

```python
median_user_ratings.head()
```

```python
min_user_ratings = ratings.transpose().fillna(round(ratings.min(axis=1), 1)).transpose().head()
```

```python
min_user_ratings.head()
```

```python
max_user_ratings = ratings.transpose().fillna(round(ratings.max(axis=1), 1)).transpose()
```

```python
max_user_ratings.head()
```

## NMF with mean user rating

```python
n_components = 20
nmf= NMF(n_components=n_components, max_iter=500)
```

```python
nmf.fit(mean_user_ratings)
```

```python
Q = pd.DataFrame(nmf.components_, columns=list(mean_user_ratings.columns), 
                 index=[f"cluster_{i+1}" for i in range(n_components)])
```

```python
Q.head(2)
```

```python
P = pd.DataFrame(nmf.transform(mean_user_ratings),
             index=mean_user_ratings.index,
            columns = [f"cluster_{i+1}" for i in range(n_components)])
```

```python
P.head(2)
```

```python
R_hat = pd.DataFrame(np.dot(P, Q), 
             index=mean_user_ratings.index,
            columns=list(mean_user_ratings.columns))
```

```python
abs(mean_user_ratings - R_hat)
```

```python
nmf.reconstruction_err_
```
